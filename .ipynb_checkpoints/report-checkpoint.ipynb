```python
import re
import matplotlib.pyplot as plt
%matplotlib inline  
file_list = ['seq-list_output.txt', 'omp-csr_output.txt',
           'seq-csr_output.txt','mpi_simple_output.txt',
            'mpi_replicated_csc_output.txt']
light_list = ['seq-list_output.txt','mpi_simple_output.txt',
            'mpi_replicated_csc_output.txt']
```

```python
def extract_teps(file_name):
    f = open(file_name,'r')
    found = re.findall('median_TEPS: (.+?)\n', f.read())
    return [float(el) for el in found]
```

```python
xlabel = range(12,21)
plt.ylabel('median_TEPS')
plt.xlabel('matrix_size(2^degree)')
for file_name in light_list:
    teps = extract_teps(file_name)
    plt.plot(xlabel,teps, label=file_name)
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
